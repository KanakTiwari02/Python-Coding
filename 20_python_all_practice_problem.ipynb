```python
# Write a program that will check whether the number is an Armstrong number or not.
num = int(input("enter the three digit no. :"))

def armstrong(num):
  num_str = str(num)
  n_digits = len(num_str)

  total = sum(int(i) ** n_digits for i in num_str)
  return total == num

if armstrong(num):
  print(f"{num} is an armstrong number")
else:
  print(f"{num} is not an armstrong number")
```

```python
text = input("Enter your input: ")

def count_chars(text):
    counts = {'Upper Case': 0, 'Lower Case': 0, 'Upper Chars': [], 'Lower Chars': []}

    for i in text:
        if i.isupper():
            counts['Upper Case'] += 1
            counts['Upper Chars'].append(i)
        elif i.islower():
            counts['Lower Case'] += 1
            counts['Lower Chars'].append(i)

    return counts

char_counts = count_chars(text)
print(char_counts)

```

```python
# Write a program to swap the key value pair for max and min values

user_input = input("Enter key-value pairs separated by commas (e.g., key1:value1, key2:value2): ").split(",")
text = {}
for i in user_input:
  key, value = i.split(":")
  key = key.strip()
  value = value.strip()
  text[key] = value

print("Original Dictionary :",text)

max_value = max(text.values())
min_value = min(text.values())

for i in text:
  if text[i] == max_value:
    print("Max Value :",text[i])
    break

for j in text:
  if text[j] == min_value:
    print("Min Value :",text[j])
    break

text[i] = min_value
text[j] = max_value
print("Swapped Dictionary",text)
```

```python
user_input = input("Enter key-value pairs separated by commas (e.g., key1:value1, key2:value2): ").split(",")
text = {}
for i in user_input:
    key, value = i.split(":")
    key = key.strip()
    value = value.strip()
    text[key] = value

print("Original Dictionary:", text)

max_value = max(text.values())
min_value = min(text.values())

max_key = min_key = None
for key, value in text.items():
    if value == max_value:
        max_key = key
    if value == min_value:
        min_key = key

text[max_key], text[min_key] = min_value, max_value

print("Swapped Dictionary:", text)
```

```python
#Write a program to find a histogram of a given set of numbers. Take bin size from the user. Print the result in the form of a dictionary.
def calculate_histogram(numbers, bin_size):
    histogram = {}

    for num in numbers:
        bin_number = num // bin_size
        if bin_number in histogram:
            histogram[bin_number] += 1
        else:
            histogram[bin_number] = 1

    return histogram

# Input numbers
numbers = list(map(int, input("Enter a set of numbers separated by spaces: ").split()))

# Input bin size
bin_size = int(input("Enter the bin size: "))

# Calculate histogram
histogram = calculate_histogram(numbers, bin_size)

# Print the histogram as a dictionary
print("Histogram:", histogram)
```

```python
# Write a program that can sort a given unsorted list. Don't use any built-in function for sorting.
unsorted_list = [12,34,45,13,32]

def bubble_sort(unsorted_list):
  N = len(unsorted_list)

  for element in range(N):
    for i in range(0, N-element-1):
      if unsorted_list[i]>unsorted_list[i+1]:
        unsorted_list[i],unsorted_list[i+1] = unsorted_list[i+1],unsorted_list[i]
bubble_sort(unsorted_list)
print(unsorted_list)


```

```python
# Write a program to check if a list is in ascending order or not
lst = [1, 3, 5, 7, 9]
def asend_order(lst):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            return False
    return True
if asend_order(lst):
    print("The list is in ascending order.")
else:
    print("The list is not in ascending order.")
```

```python
# Write a program to calculate the sum of the following series till the nth term
# 1/1! + 2/2! + 3/3! + 4/4! +.+ n/n!
n = int(input("Enter the value of n: "))
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def series_sum(n):
    series_sum = 0
    for i in range(1, n + 1):
        term = i / factorial(i)
        series_sum += term
    return series_sum


result = series_sum(n)
print("Sum of the series up to the", n, "term:", result)
```

```python
# # Write a Python Program to Find the Sum of the Series till the nth term:
# 1 + x^2/2 + x^3/3 +  x^n/n
# n will be provided by the user
x = float(input("Enter the value of x: "))
n = int(input("Enter the value of n: "))

def series_sum(x, n):
    series_sum = 0
    for i in range(1, n + 1):
        term = (x ** i) / i
        series_sum += term
    return series_sum

result = series_sum(x, n)
print("Sum of the series up to the", n, "term:", result)
```

```python
# Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.
n = int(input("Enter the value of n: "))
add = 0
count = 0
avg = 0

while True:
  if n != 0:
    add = add + n
    count += 1
    avg = add/count
    n = int(input("enter another no."))
  else:
    print("thank you")
print("sum :",add)
print("average :",avg)
print("no of attempts :", count)
```
